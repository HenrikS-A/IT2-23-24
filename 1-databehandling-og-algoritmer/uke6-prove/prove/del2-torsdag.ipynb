```python
# 1. Åpne filen `pokemon.json` og print ordboken til `Bulbasaur` (den første i listen)

import json

with open("pokemon.json", encoding="utf-8") as fil:
    pokemons = json.load(fil)

print(pokemons[0])
```

```python
# 2. Print det engelske navnet på Bulbasaur, HP, attack og defense 
#    i terminalen på én linje.
#    Slik: Bulbasaur - HP: 45, Attack: 49, Defense: 49

print(f"{pokemons[0]['name']['english']} - HP: {pokemons[0]['base']['HP']}, Attack: {pokemons[0]['base']['Attack']}, Defense: {pokemons[0]['base']['Defense']}")
```

```python
# 3. Print det engelske navnet, HP, attack og defense på alle pokemon-ene i filen.
#    De skal printes på samme måte som i deloppgave 2

for pokemon in pokemons:
    print(f"{pokemon['name']['english']} - HP: {pokemon['base']['HP']}, Attack: {pokemon['base']['Attack']}, Defense: {pokemon['base']['Defense']}")
    
```

> OM OPPGAVE 4  
> Velg enten a. eller b.  
> a. gir mest uttelling.

```python
# VELG ENTEN 4a. ELLER 4b.
# 4a. Lag en ordbok som teller hvor mange Pokemon som er av de forskjellige typene.

typer_pokemons = {}
for pokemon in pokemons:
    for typen in pokemon["type"]:
        if typen not in typer_pokemons:
            typer_pokemons[typen] = 1
        else:
            typer_pokemons[typen] += 1

# print(typer_pokemons)
```

```python
# VELG ENTEN 4a. ELLER 4b.
# 4b. Print alle typene til hver pokemeon, en type per linje. En type skal printes selv 
#     om den har blitt printet før.
# Eks:
# Grass
# Poison
# Grass
# Poison
# ...

for pokemon in pokemons:
    for typen in pokemon["type"]:
        print(typen)
```

```python
# 5. Tegn et stolpediagram med antall nominerte gruppert etter produsent

import matplotlib.pyplot as plt

prod_nominert = {}
for pris in humorprisen_2023:
    for nominert in pris["nominerte"]:
        if nominert["produsent"] not in prod_nominert:
            prod_nominert[nominert["produsent"]] = 1
        else:
            prod_nominert[nominert["produsent"]] += 1

# Sorterert i synkende rekkefølge
prod_nominert_sortert = sorted(prod_nominert.items(), key=lambda produsent: produsent[1], reverse=True)

produsenter = [prod[0] for prod in prod_nominert_sortert]
nominerte = [prod[1] for prod in prod_nominert_sortert]

plt.bar(produsenter, nominerte)

## Verdier over søylene
for i, antall in enumerate(nominerte):
    plt.text(i, antall + 0.05, str(antall), horizontalalignment="center")

plt.xticks(rotation=20) # Roterer teksten til x_verdiene litt
plt.title("Antall nominerte fra hver produsent")
plt.xlabel("Produsent")
plt.ylabel("Antall nominerte")
plt.show()
```
