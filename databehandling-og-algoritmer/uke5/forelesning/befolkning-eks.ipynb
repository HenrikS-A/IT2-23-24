## Eksempel med plotting fra json-fil

```python
import json
import matplotlib.pyplot as plt

with open("land.json", encoding="utf-8") as fil:
    data = json.load(fil)

data[0] # Sjekker at alt gikk fint med innlesning av fila
```

```python
verdensdeler = {}

for land in data:
    if land["verdensdel"] not in verdensdeler:
        verdensdeler[land["verdensdel"]] = 1
    else:
        verdensdeler[land["verdensdel"]] += 1

print(verdensdeler)
```

```python
x_verdier = list(verdensdeler.keys())
y_verdier = list(verdensdeler.values())

plt.bar(x_verdier, y_verdier)


# ## Verdier over søylene
for i, verdien in enumerate(y_verdier):
    plt.text(i, verdien + 0.1, str(verdien), horizontalalignment="center")
# ## Diverse utforming
plt.title("Antall land på topplista")
plt.xlabel("Verdensdeler")
plt.ylabel("Antall land")

plt.style.use("fivethirtyeight")

plt.show()
```
